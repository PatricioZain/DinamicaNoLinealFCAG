```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi
```

```python
K=1
nc = int(100)
n = int(1e4)
x0 = 0
#solo si nc = 1
y0 = 0
```

```python
nc = int(nc)
n = int(n)
K2=K/(2.*pi)
y0 = np.linspace(0,1,nc)
```

```python
xx = np.zeros(n)
yy = np.zeros(n)


for j in range (nc): #Por cada condición inicial
	x = x0
	
	if(nc==1):
		y = y0
	else:
		y = y0[j]
		
	for i in range(n):	#Cada iteración
		
		y = y + K2 * np.sin(2.*pi*x)
		x = x + y 
		
		#con la funcion mod(,1) garantizo que x e y sean menores que 1	

		y = np.mod(y,1)
		x = np.mod(x,1)
		
		yy[i] = y
		xx[i] = x
		
	plt.plot(xx,yy,'.',markersize='0.01',color='k')		

plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0,1)
plt.ylim(0,1)
#plt.title(nombre)		
plt.show()

```
